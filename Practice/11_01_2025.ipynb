### Write a while loop that keeps asking the user for a number until they enter a negative number. At the end, print the sum of all non-negative numbers entered.

```python
total = 0
while True:
    num = int(input("Number: "))
    if num <0:
        break
    total += num
    print(total)
```

### Using a while loop, print the first 10 Fibonacci numbers (starting from 0, 1).

```python
second_last_num = 0
last_num = 1
count = 0
while count <10:
    print(second_last_num, end = ', ')
    second_last_num, last_num = last_num, second_last_num + last_num
    count += 1
```

```python
# second solution
fib = [0, 1]
while len(fib) < 10:
    fib.append(fib[-1] + fib[-2])
print(fib)
```

### You have a starting number n = 987654. Using a while loop, count how many times you can divide n by 2 (integer division) until it becomes 0.

```python
n = 987654
steps = 0
while n > 0:
    n//=2
    steps += 1
print (steps)
```

### You have a password "python123". Let the user try up to 3 times to enter the correct password using a while loop. Print “Access granted” or “Blocked”.

```python
password = 'python123'
number_of_tries = 0
while number_of_tries < 3:
    entered_password = input('Please enter your password')
    if entered_password == password:
        print("Access granted!")
        break
    else:
        print('Incorrect password, please try again!')
    number_of_tries += 1

    if number_of_tries == 3:
        print('You are blocked! Please ccontact the customer service')
```
